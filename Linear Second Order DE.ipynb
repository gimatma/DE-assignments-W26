## The set up

The first step is to import the required libraries

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

Since we are doing symbolic calculations, we must define the variables

```python
x = sp.symbols('x')             # Name the independent variable
y = sp.Function('y')            # Make the dependent variable a function
```

A initial value problem for a second order equation has two initial conditions. To evaluate them, we first need to define the derivatives 

```python
yprime =  sp.diff( y(x), x )    # The first derivative of y with respect to x
y2prime = sp.diff( y(x), x, x ) # The second derivative of y with respect to x twice
```

Finally, we can define up front all the parameters needed to construct the equation. Remember that we are considering an equation of the form $ay''+by'+cy=f(x)$

So, we need to set up the constants $a$, $b$, $c$, the function $f(x)$ and the initial data $(x_0,y_0,y_1)$

```python
# The constant coefficients 
a = 1                  #(default a = 1)
b = 2                  #(default b = 2) 
c = 3                  #(default c = 3)

# The initial data
x0 = 0                 #Initial value of x  (default x0 = 0)
y0 = 1                 #Initial value of y  (default y(x0) = 1)
y1= -2                 #Initial value of y' (default y'(x0) = -2)

```

Here is the definition of the forcing term $f$. As default, it is set to $f=0$. To consider the nonhomogeneous equation, just delete the `#` in front of the `omega` and the `f` variables 

```python
# The forcing term f. 
f=0

# Uncomment to consider the nonhomogeneous equations. Be careful with the indent 
# (there can't be any space before the variables)

#omega = sp.sqrt(c/a)  
#f=sp.cos(omega*x)
```

Finally, these are some plotting parameters

```python
xmin = 0               # The initial and final point of the x-axis
xmax = 10              # Feel free to modify as needed

autoYaxes = 1          #Set to 1 to auto-scale, or 0 to use the manual values
ymin = -1              # The initial and final point of the x-axis 
ymax = 3
```

### Review of the parameters

Let's print some key quantities to help the theoretical analysis of the solution. Notice that in SymPy the imaginary unit is represented by `I`. 

```python
D = b**2 - 4*a*c                   #The discriminant - what does this tell you?

lambda1 = -b/2/a + sp.sqrt(D)/(2*a)  # The eigenvalues
lambda2 = -b/2/a - sp.sqrt(D)/(2*a)  

# We can treat these as: lambda = alpha + i(beta)
alpha = -b/(2*a)                     # Negative alpha gives exponential decay
beta = sp.sqrt(-D)/(2*a)             # Note how I'm defining this using -D to account for the i, and the positive square root

print( "Discriminant =",D,"\n\n"+"Eigenvalues:",
      "\n\t"+"lambda1 =",lambda1,"\n\t"+"lambda2 =",lambda2,"\n" )
```

## Actually finding the solution 

Recall that to find the solution, we must first define the differential equation

```python
# The values on the left hand side of the equal sign
LHS = a*y2prime + b*yprime + c*y(x)  

# This is the equation. It makes LHS=f
DE = sp.Eq(LHS, f )  

print("The SOLDE is:")
sp.pprint(DE)
```

To get the solution, we use `dsolve`. Note that there is a slightly different syntax to account for the two initial conditions. 

```python
ivp = sp.dsolve( DE, y(x), ics={ y(x0):y0 , #These are the initial conditions for y...
                                y(x).diff(x).subs(x, x0): y1} )    #...and for y' 


from sympy import init_printing
init_printing()

print("\n"+"When the initial conditions are \n\t",
      "y("+str(x0)+") =",y0,"and \n\t",
      "y'("+str(x0)+") =",y1,"\n"+ 
      "the solution is:\n")

ivp
```

## The plot

We can plot the solution directly, but converting the solution to a numerical array, lets us find the max and min values to use them as plotting limits

```python
ysol= ivp.rhs   #Pulling out the solution above

# This is the extra y-axis control

Ns = 100*xmax  
s = np.linspace(xmin, xmax, Ns + 1)    #Here's the numerical x-axis array
y = sp.lambdify(x, ysol, 'numpy')      #Turning the solution into a function

if autoYaxes == 1:
    ymin = np.min( y(s) ) - 1          #Creating a numerical array y(s) and... 
    ymax = np.max( y(s) ) + 1          #...looking for its max and min values
```

Now the commands to create the plot

```python
sp.plot(ysol, 
            (x, xmin, xmax),     
             ylim = (ymin,ymax),  
             xlabel = 'x',        
             ylabel = 'y(x)')

plt.show()
```
